```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True, color="skyblue")
plt.title("총지불금액")
plt.show()
```

```python
#tips 데이터셋을 이용하여 요일(day)별로 손님이 남긴 팁(tip)의 평균을 구하고, barplot으로 시각화하세요.

sns.barplot(x = 'day' , y = 'tip', data = tips)
plt.title("요일별 지불금액")
plt.show()
```

```python
# 흡연 여부에 따른 총지불금액 분포
tips['smoker_label'] = tips['smoker'].map({'Yes': '흡연', 'No': '비흡연'})
sns.scatterplot(x = "smoker_label", y = "total_bill", data = tips)

plt.title("흡연 여부에 따른 총 지불금액 분포")
```

```python
#팁 비율(tip/total_bill)을 계산하여 새로운 컬럼(tip_pct)을 만들고, 성별(sex)에 따른 평균 팁 비율을 barplot으로 시각화하세요
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
sns.barplot(x = 'sex', y = 'tip_pct', data = tips)
```

```python
#시간대별 총지불금액과 팁의 관계
#점심(Lunch)과 저녁(Dinner) 시간대(time)별로 총지불금액(total_bill)과 팁(tip)의 관계를 산점도(scatterplot)로 시각화하세요.
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'time' , data= tips )
plt.title("시간대별 총지불금액과 팁의 관계")
```

```python
#인원수에 따른 평균 총지불금액
#식사 인원수(size)별로 평균 총지불금액(total_bill)을 구하고, barplot으로 시각화하세요.
```

```python
sns.barplot(x='size', y='total_bill', data=tips, estimator=lambda x: sum(x)/len(x))
plt.title('식사 인원수별 평균 총지불금액')
plt.xlabel('식사 인원수')
plt.ylabel('평균 총지불금액')
plt.show()
```

# flights

```python
flights = sns.load_dataset('flights')
```

```python
from IPython.display import display, HTML


# HTML 테이블로 변환 후 스크롤 영역 지정
html = flights.to_html()
scrollable_html = f"""
<div style='height:400px; overflow:auto;'>
{html}
</div>
"""

display(HTML(scrollable_html))
```

```python
#연도별 항공 승객 수 변화
#연도(year)별로 항공 승객 수(passengers)의 합계를 구하고, 선 그래프(line plot)로 시각화하세요.
```

```python
sns.lineplot( x= 'year' , y = 'passengers' , data = flights)
```

```python
#2. 월별 평균 승객 수
#월(month)별로 평균 승객 수를 구하고, barplot으로 시각화하세요
```

```python
import numpy as np
sns.barplot( x = 'month', y = 'passengers' ,data = flights, estimator = np.mean)
```

```python
#피벗 테이블과 히트맵(heatmap)
#flights 데이터셋을 연도(year)와 월(month)을 행과 열로 하여 승객 수(passengers)의 합계를 나타내는 피벗 테이블을 만들고, 이를 히트맵으로 시각화하세요
```

```python
flights_pivot = flights.pivot_table(
    index='month',
    columns='year',
    values='passengers',
    aggfunc='sum'
)
print(flights_pivot)
```

```python
sns.heatmap(
    flights_pivot,
    annot=True,    # 각 셀에 숫자 표시
    fmt='d',       # 정수 형식
)

plt.title('연도별 월간 항공사 이용 승객 수')
plt.xlabel('연도')
plt.ylabel('월')
plt.show()
```

```python
#승객 수가 가장 많았던 연도와 월 찾기
#승객 수(passengers)가 가장 많았던 연도(year)와 월(month)을 구하세요.
```

```python
max_flight = flights.loc[flights['passengers'].idxmax()]
print(max_flight)
```

```python
#1950년대와 1960년대의 승객 수 비교
#1950년대와 1960년대의 연평균 승객 수를 각각 구하고, barplot으로 시각화하세요
```

```python
import pandas as pd

year_1950 = flights[(flights['year'] >= 1950) & (flights['year'] < 1960)]

# 1960년대 데이터 추출 (1960-1969)
year_1960 = flights[(flights['year'] >= 1960) & (flights['year'] < 1970)]

# 연평균 계산 (각 년도 평균의 평균)
avg_1950s = year_1950.groupby('year')['passengers'].mean().mean()
avg_1960s = year_1960.groupby('year')['passengers'].mean().mean()


# 결과 테이블 생성
avg_passengers = pd.DataFrame({
    'Decade': ['1950s', '1960s'],
    'Average Passengers': [avg_1950s, avg_1960s]
})

print(avg_passengers)
# 시각화
plt.figure(figsize=(8,6))
sns.barplot(x='Decade', y='Average Passengers', data=avg_passengers, palette='pastel')
plt.title('1950년대 vs 1960년대 연평균 승객 수 비교')
plt.ylabel('연평균 승객 수 (명)')
plt.show()
```
