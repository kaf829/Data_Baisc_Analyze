```python
import pandas as pd


file_path = "./yeobo_final_temperature_filled.csv"
df = pd.read_csv(file_path)

df.info(), df.head()
```

```python
df["일시"] = pd.to_datetime(df["일시"])
df = df.sort_values("일시").reset_index(drop=True)

# 평균기온 결측치 제거
df_temp = df[["일시", "평균기온(°C)"]].dropna()

# T+1 예측을 위해 평균기온 시프트 생성
df_temp["다음날_기온"] = df_temp["평균기온(°C)"].shift(-1)

# 마지막 행은 다음날 기온이 없으므로 제거
df_temp = df_temp.dropna()

# 결과 확인
df_temp.head()
```

```python
for i in range(1, 4):
    df_temp[f"lag_{i}"] = df_temp["평균기온(°C)"].shift(i)

df_temp = df_temp.dropna()

feature_cols = ["lag_1", "lag_2", "lag_3"]
X = df_temp[feature_cols]
y = df_temp["다음날_기온"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, r2
```

```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt


df = pd.read_csv("yeobo.csv")
df["일시"] = pd.to_datetime(df["일시"])
df = df.sort_values("일시")
df = df[["지점명", "일시", "평균기온(°C)"]].dropna()


results = []
accuracy_scores = []


for station in df["지점명"].unique():
    station_df = df[df["지점명"] == station][["일시", "평균기온(°C)"]].copy()
    station_df = station_df.rename(columns={"일시": "ds", "평균기온(°C)": "y"})

    model = Prophet()
    model.fit(station_df)

    future = model.make_future_dataframe(periods=700)
    forecast = model.predict(future)


    forecast_filtered = forecast[(forecast["ds"] >= "2026-09-01") & (forecast["ds"] < "2026-12-01")]
    forecast_filtered["지점명"] = station
    forecast_filtered = forecast_filtered[["ds", "yhat", "지점명"]].rename(columns={"ds": "날짜", "yhat": "예측 평균기온"})
    results.append(forecast_filtered)


    if len(station_df) > 365:
        past = station_df[-365:]
        train = station_df[:-365]

        model_eval = Prophet()
        model_eval.fit(train)
        eval_future = model_eval.make_future_dataframe(periods=365)
        eval_forecast = model_eval.predict(eval_future)

        actual = past["y"].values
        predicted = eval_forecast[-365:]["yhat"].values

        if len(actual) == len(predicted):
            mse = mean_squared_error(actual, predicted)
            r2 = r2_score(actual, predicted)
            accuracy_scores.append({"지점명": station, "MSE": round(mse, 2), "R2": round(r2, 3)})
    

forecast_all = pd.concat(results, ignore_index=True)
accuracy_df = pd.DataFrame(accuracy_scores)

# 5. 예측 결과 출력
print("\n📅 [2026년 9~11월 평균기온 예측 결과]")
print(forecast_all)

print("\n📊 [지점별 Prophet 예측 정확도 평가 (최근 1년 기준)]")
print(accuracy_df)

# 6. 📈 지점별 시각화
import matplotlib.dates as mdates

for station in forecast_all["지점명"].unique():
    data = forecast_all[forecast_all["지점명"] == station]

    plt.figure(figsize=(12, 4))
    plt.rcParams["font.family"] = "Malgun Gothic"
    plt.rcParams['axes.unicode_minus'] = False
    plt.plot(data["날짜"], data["예측 평균기온"], marker='o', linestyle='-', label=f"{station}")
    plt.title(f"2026년 가을 평균기온 예측 - {station}")
    plt.xlabel("날짜")
    plt.ylabel("예측 평균기온 (°C)")
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# 1. 데이터 불러오기 및 전처리
df = pd.read_csv("yeobo.csv")
df["ds"] = pd.to_datetime(df["일시"])
df["y"] = df["평균기온(°C)"]
df["humidity"] = df["평균 상대습도(%)"]
df = df[["ds", "y", "humidity"]].dropna()

# 2. Prophet 모델 정의 및 보조변수 등록
model = Prophet()
model.add_regressor("humidity")  # ✅ 습도 추가

# 3. 모델 학습
model.fit(df)

# 4. 미래 예측 날짜 생성 (365일 예측)
future = model.make_future_dataframe(periods=365)

# 5. 미래 습도값도 넣어야 예측 가능
# 예: 최근 1년간 습도값을 순환해서 채움 (간단한 예)
recent_humidity = df["humidity"].iloc[-365:].values.tolist()
repeat_count = len(future) // len(recent_humidity) + 1
future["humidity"] = (recent_humidity * repeat_count)[:len(future)]

# 6. 예측 수행
forecast = model.predict(future)

# 7. 시각화
plt.figure(figsize=(12, 5))
plt.plot(forecast["ds"], forecast["yhat"], label="예측 평균기온")
plt.title("Prophet 예측 결과 (습도 포함)")
plt.xlabel("날짜")
plt.ylabel("평균기온 (°C)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 8. 습도 영향 분석
model.plot_components(forecast)
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 데이터 로딩 및 전처리
df = pd.read_csv("yeobo.csv")
df["ds"] = pd.to_datetime(df["일시"])
df["y"] = df["평균기온(°C)"]
df["humidity"] = df["평균 상대습도(%)"]
df = df[["지점명", "ds", "y", "humidity"]].dropna()

for station in df["지점명"].unique():
    station_df = df[df["지점명"] == station][["ds", "y", "humidity"]]
    if len(station_df) < 400:
        continue

    # Prophet 모델 정의
    model = Prophet()
    model.add_regressor("humidity")
    model.fit(station_df)

    # 2026년 예측용 future 생성
    future = pd.date_range(start="2026-01-01", end="2026-12-31", freq="D")
    future_df = pd.DataFrame({"ds": future})

    # 미래 습도 보간 또는 순환 사용
    recent_humidity = station_df["humidity"].iloc[-365:].tolist()
    future_df["humidity"] = (recent_humidity * ((len(future_df) // len(recent_humidity)) + 1))[:len(future_df)]

    # 예측 수행
    forecast = model.predict(future_df)

    # 🎯 2026년 9~11월만 추출
    fall_forecast = forecast[(forecast["ds"] >= "2026-09-01") & (forecast["ds"] < "2026-12-01")]

    # 시각화
    plt.figure(figsize=(12, 4))
    plt.plot(fall_forecast["ds"], fall_forecast["yhat"], label="예측 평균기온")
    plt.fill_between(fall_forecast["ds"], fall_forecast["yhat_lower"], fall_forecast["yhat_upper"], alpha=0.3, label="신뢰 구간")
    plt.title(f"{station} - 2026년 가을 평균기온 예측 (습도 포함)")
    plt.xlabel("날짜")
    plt.ylabel("예측 평균기온 (°C)")
    plt.grid(True)
    plt.legend()
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np


file_path = "yeobo_final_temperature_filled.csv"
df = pd.read_csv(file_path)
df["일시"] = pd.to_datetime(df["일시"])
df = df[["지점명", "일시", "평균기온(°C)", "평균 상대습도(%)"]].dropna()

# 날짜 특징 추출
df["year"] = df["일시"].dt.year
df["month"] = df["일시"].dt.month
df["day"] = df["일시"].dt.day
df["dayofyear"] = df["일시"].dt.dayofyear

# 예측 대상 날짜 생성 (2026년 9~11월)
future_dates = pd.date_range(start="2026-09-01", end="2026-11-30", freq="D")
future_df = pd.DataFrame({"일시": future_dates})
future_df["year"] = future_df["일시"].dt.year
future_df["month"] = future_df["일시"].dt.month
future_df["day"] = future_df["일시"].dt.day
future_df["dayofyear"] = future_df["일시"].dt.dayofyear 

results = []
accuracy_scores = []

for station in df["지점명"].unique():
    station_df = df[df["지점명"] == station].copy()
    
    if len(station_df) < 3650:
        continue

    # 학습용 데이터: 2015~2024
    train_df = station_df[(station_df["일시"] >= "2015-01-01") & (station_df["일시"] <= "2024-12-31")].copy()
    test_df = station_df[(station_df["일시"] >= "2024-09-01") & (station_df["일시"] < "2024-12-01")].copy()

    # 입력 및 출력 설정
    X_train = train_df[["month", "day", "dayofyear", "평균 상대습도(%)"]]
    y_train = train_df["평균기온(°C)"]
    X_test = test_df[["month", "day", "dayofyear", "평균 상대습도(%)"]]
    y_test = test_df["평균기온(°C)"]

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # 정확도 측정
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    accuracy_scores.append({"지점명": station, "MSE": round(mse, 2), "R2": round(r2, 3)})

    # 미래 예측용 데이터 생성
    future_df_copy = future_df.copy()
    avg_humidity = station_df["평균 상대습도(%)"].mean()
    future_df_copy["평균 상대습도(%)"] = avg_humidity

    X_future = future_df_copy[["month", "day", "dayofyear", "평균 상대습도(%)"]]
    y_future_pred = model.predict(X_future)

    result = pd.DataFrame({
        "날짜": future_df_copy["일시"],
        "예측 평균기온": y_future_pred,
        "지점명": station
    })
    results.append(result)

# 전체 결과 병합
forecast_all = pd.concat(results, ignore_index=True)
accuracy_df = pd.DataFrame(accuracy_scores)

print(accuracy_df.head())
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np


file_path = "yeobo_final_temperature_filled.csv"
df = pd.read_csv(file_path)
df["일시"] = pd.to_datetime(df["일시"])
df = df[["지점명", "일시", "평균기온(°C)", "평균 상대습도(%)"]].dropna()

# 날짜 특징 추출
df["year"] = df["일시"].dt.year
df["month"] = df["일시"].dt.month
df["day"] = df["일시"].dt.day
df["dayofyear"] = df["일시"].dt.dayofyear

# 예측 대상 날짜 생성 (2026년 9~11월)
future_dates = pd.date_range(start="2026-09-01", end="2026-11-30", freq="D")
future_df = pd.DataFrame({"일시": future_dates})
future_df["year"] = future_df["일시"].dt.year
future_df["month"] = future_df["일시"].dt.month
future_df["day"] = future_df["일시"].dt.day
future_df["dayofyear"] = future_df["일시"].dt.dayofyear

results = []
accuracy_scores = []

for station in df["지점명"].unique():
    station_df = df[df["지점명"] == station].copy()
    
    if len(station_df) < 3650:
        continue

    # 학습용 데이터: 2015~2024
    train_df = station_df[(station_df["일시"] >= "2015-01-01") & (station_df["일시"] <= "2024-12-31")].copy()
    test_df = station_df[(station_df["일시"] >= "2024-09-01") & (station_df["일시"] < "2024-12-01")].copy()

    # 입력 및 출력 설정
    X_train = train_df[["month", "day", "dayofyear", "평균 상대습도(%)"]]
    y_train = train_df["평균기온(°C)"]
    X_test = test_df[["month", "day", "dayofyear", "평균 상대습도(%)"]]
    y_test = test_df["평균기온(°C)"]

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # 정확도 측정
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    accuracy_scores.append({"지점명": station, "MSE": round(mse, 2), "R2": round(r2, 3)})

    # 미래 예측용 데이터 생성
    future_df_copy = future_df.copy()
    avg_humidity = station_df["평균 상대습도(%)"].mean()
    future_df_copy["평균 상대습도(%)"] = avg_humidity

    X_future = future_df_copy[["month", "day", "dayofyear", "평균 상대습도(%)"]]
    y_future_pred = model.predict(X_future)

    result = pd.DataFrame({
        "날짜": future_df_copy["일시"],
        "예측 평균기온": y_future_pred,
        "지점명": station
    })
    results.append(result)

# 전체 결과 병합
forecast_all = pd.concat(results, ignore_index=True)
accuracy_df = pd.DataFrame(accuracy_scores)

print(accuracy_df.head())
```

```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv("./yeobo_final_temperature_filled.csv")
df["일시"] = pd.to_datetime(df["일시"])
df["평균기온(°C)"] = pd.to_numeric(df["평균기온(°C)"], errors="coerce")
df = df[["지점명", "일시", "평균기온(°C)"]].dropna()

results = []
scores = []

for station in df["지점명"].unique():
    station_df = df[df["지점명"] == station][["일시", "평균기온(°C)"]].copy()
    station_df["일시"] = pd.to_datetime(station_df["일시"])
    station_df["평균기온(°C)"] = pd.to_numeric(station_df["평균기온(°C)"], errors="coerce")
    station_df = station_df.set_index("일시").resample("D").mean().ffill()
    
    station_df["target"] = station_df["평균기온(°C)"].shift(-1)
    for i in range(1, 4):
        station_df[f"lag_{i}"] = station_df["평균기온(°C)"].shift(i)

    station_df = station_df.dropna()

    X = station_df[["lag_1", "lag_2", "lag_3"]]
    y = station_df["target"]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    X_train, y_train = X_scaled[:-90], y[:-90]
    X_test, y_test = X_scaled[-90:], y[-90:]
    test_dates = y_test.index

    model = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    pred_df = pd.DataFrame({
        "날짜": test_dates,
        "예측 평균기온": y_pred,
        "실제 평균기온": y_test.values,
        "지점명": station
    })
    results.append(pred_df)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    scores.append({"지점명": station, "MSE": round(mse, 2), "R2": round(r2, 3)})

forecast_xgb = pd.concat(results, ignore_index=True)
accuracy_xgb = pd.DataFrame(scores)

accuracy_xgb.head()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
import numpy as np

# 데이터 불러오기
file_path = "weather_all_data_filled_mice.csv"
df = pd.read_csv(file_path)
df["일시"] = pd.to_datetime(df["일시"])

# 필요한 컬럼만 선택 & 결측치 제거
df = df[["지점명", "일시", "평균기온(°C)", "평균 지면온도(°C)"]].dropna()

# 날짜 관련 특징 추출
df["year"] = df["일시"].dt.year
df["month"] = df["일시"].dt.month
df["day"] = df["일시"].dt.day
df["dayofyear"] = df["일시"].dt.dayofyear

# 예측할 미래 날짜 (2026년 9~11월)
future_dates = pd.date_range(start="2026-09-01", end="2026-11-30", freq="D")
future_df = pd.DataFrame({"일시": future_dates})
future_df["year"] = future_df["일시"].dt.year
future_df["month"] = future_df["일시"].dt.month
future_df["day"] = future_df["일시"].dt.day
future_df["dayofyear"] = future_df["일시"].dt.dayofyear 

results = []
accuracy_scores = []

for station in df["지점명"].unique():
    station_df = df[df["지점명"] == station].copy()
    
    if len(station_df) < 3650:
        continue

    # 학습 데이터 분리
    train_df = station_df[(station_df["일시"] >= "2015-01-01") & (station_df["일시"] <= "2024-12-31")].copy()
    test_df = station_df[(station_df["일시"] >= "2024-09-01") & (station_df["일시"] < "2024-12-01")].copy()

    # 입력(X), 출력(y) 정의
    X_train = train_df[["month", "day", "dayofyear", "평균 지면온도(°C)"]]
    y_train = train_df["평균기온(°C)"]
    X_test = test_df[["month", "day", "dayofyear", "평균 지면온도(°C)"]]
    y_test = test_df["평균기온(°C)"]

    # 결측치 처리
    imputer = SimpleImputer(strategy="mean")
    X_train = imputer.fit_transform(X_train)
    X_test = imputer.transform(X_test)

    # 모델 학습
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # 정확도 측정
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    accuracy_scores.append({"지점명": station, "MSE": round(mse, 2), "R2": round(r2, 3)})

    # 미래 예측
    future_df_copy = future_df.copy()
    avg_ground_temp = station_df["평균 지면온도(°C)"].mean()
    future_df_copy["평균 지면온도(°C)"] = avg_ground_temp

    X_future = future_df_copy[["month", "day", "dayofyear", "평균 지면온도(°C)"]]
    X_future = imputer.transform(X_future)
    y_future_pred = model.predict(X_future)

    result = pd.DataFrame({
        "날짜": future_df_copy["일시"],
        "예측 평균기온": y_future_pred,
        "지점명": station
    })
    results.append(result)

# 전체 결과 병합
forecast_all = pd.concat(results, ignore_index=True)
accuracy_df = pd.DataFrame(accuracy_scores)


print(forecast_all)
print(accuracy_df.head(5))
```
