```python
import pandas as pd


file_path = "./yeobo_final_temperature_filled.csv"
df = pd.read_csv(file_path)

df.info(), df.head()
```

```python
df["ÏùºÏãú"] = pd.to_datetime(df["ÏùºÏãú"])
df = df.sort_values("ÏùºÏãú").reset_index(drop=True)

# ÌèâÍ∑†Í∏∞Ïò® Í≤∞Ï∏°Ïπò Ï†úÍ±∞
df_temp = df[["ÏùºÏãú", "ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"]].dropna()

# T+1 ÏòàÏ∏°ÏùÑ ÏúÑÌï¥ ÌèâÍ∑†Í∏∞Ïò® ÏãúÌîÑÌä∏ ÏÉùÏÑ±
df_temp["Îã§ÏùåÎÇ†_Í∏∞Ïò®"] = df_temp["ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"].shift(-1)

# ÎßàÏßÄÎßâ ÌñâÏùÄ Îã§ÏùåÎÇ† Í∏∞Ïò®Ïù¥ ÏóÜÏúºÎØÄÎ°ú Ï†úÍ±∞
df_temp = df_temp.dropna()

# Í≤∞Í≥º ÌôïÏù∏
df_temp.head()
```

```python
for i in range(1, 4):
    df_temp[f"lag_{i}"] = df_temp["ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"].shift(i)

df_temp = df_temp.dropna()

feature_cols = ["lag_1", "lag_2", "lag_3"]
X = df_temp[feature_cols]
y = df_temp["Îã§ÏùåÎÇ†_Í∏∞Ïò®"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, r2
```

```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt


df = pd.read_csv("yeobo.csv")
df["ÏùºÏãú"] = pd.to_datetime(df["ÏùºÏãú"])
df = df.sort_values("ÏùºÏãú")
df = df[["ÏßÄÏ†êÎ™Ö", "ÏùºÏãú", "ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"]].dropna()


results = []
accuracy_scores = []


for station in df["ÏßÄÏ†êÎ™Ö"].unique():
    station_df = df[df["ÏßÄÏ†êÎ™Ö"] == station][["ÏùºÏãú", "ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"]].copy()
    station_df = station_df.rename(columns={"ÏùºÏãú": "ds", "ÌèâÍ∑†Í∏∞Ïò®(¬∞C)": "y"})

    model = Prophet()
    model.fit(station_df)

    future = model.make_future_dataframe(periods=700)
    forecast = model.predict(future)


    forecast_filtered = forecast[(forecast["ds"] >= "2026-09-01") & (forecast["ds"] < "2026-12-01")]
    forecast_filtered["ÏßÄÏ†êÎ™Ö"] = station
    forecast_filtered = forecast_filtered[["ds", "yhat", "ÏßÄÏ†êÎ™Ö"]].rename(columns={"ds": "ÎÇ†Ïßú", "yhat": "ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò®"})
    results.append(forecast_filtered)


    if len(station_df) > 365:
        past = station_df[-365:]
        train = station_df[:-365]

        model_eval = Prophet()
        model_eval.fit(train)
        eval_future = model_eval.make_future_dataframe(periods=365)
        eval_forecast = model_eval.predict(eval_future)

        actual = past["y"].values
        predicted = eval_forecast[-365:]["yhat"].values

        if len(actual) == len(predicted):
            mse = mean_squared_error(actual, predicted)
            r2 = r2_score(actual, predicted)
            accuracy_scores.append({"ÏßÄÏ†êÎ™Ö": station, "MSE": round(mse, 2), "R2": round(r2, 3)})
    

forecast_all = pd.concat(results, ignore_index=True)
accuracy_df = pd.DataFrame(accuracy_scores)

# 5. ÏòàÏ∏° Í≤∞Í≥º Ï∂úÎ†•
print("\nüìÖ [2026ÎÖÑ 9~11Ïõî ÌèâÍ∑†Í∏∞Ïò® ÏòàÏ∏° Í≤∞Í≥º]")
print(forecast_all)

print("\nüìä [ÏßÄÏ†êÎ≥Ñ Prophet ÏòàÏ∏° Ï†ïÌôïÎèÑ ÌèâÍ∞Ä (ÏµúÍ∑º 1ÎÖÑ Í∏∞Ï§Ä)]")
print(accuracy_df)

# 6. üìà ÏßÄÏ†êÎ≥Ñ ÏãúÍ∞ÅÌôî
import matplotlib.dates as mdates

for station in forecast_all["ÏßÄÏ†êÎ™Ö"].unique():
    data = forecast_all[forecast_all["ÏßÄÏ†êÎ™Ö"] == station]

    plt.figure(figsize=(12, 4))
    plt.rcParams["font.family"] = "Malgun Gothic"
    plt.rcParams['axes.unicode_minus'] = False
    plt.plot(data["ÎÇ†Ïßú"], data["ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò®"], marker='o', linestyle='-', label=f"{station}")
    plt.title(f"2026ÎÖÑ Í∞ÄÏùÑ ÌèâÍ∑†Í∏∞Ïò® ÏòàÏ∏° - {station}")
    plt.xlabel("ÎÇ†Ïßú")
    plt.ylabel("ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò® (¬∞C)")
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ Î∞è Ï†ÑÏ≤òÎ¶¨
df = pd.read_csv("yeobo.csv")
df["ds"] = pd.to_datetime(df["ÏùºÏãú"])
df["y"] = df["ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"]
df["humidity"] = df["ÌèâÍ∑† ÏÉÅÎåÄÏäµÎèÑ(%)"]
df = df[["ds", "y", "humidity"]].dropna()

# 2. Prophet Î™®Îç∏ Ï†ïÏùò Î∞è Î≥¥Ï°∞Î≥ÄÏàò Îì±Î°ù
model = Prophet()
model.add_regressor("humidity")  # ‚úÖ ÏäµÎèÑ Ï∂îÍ∞Ä

# 3. Î™®Îç∏ ÌïôÏäµ
model.fit(df)

# 4. ÎØ∏Îûò ÏòàÏ∏° ÎÇ†Ïßú ÏÉùÏÑ± (365Ïùº ÏòàÏ∏°)
future = model.make_future_dataframe(periods=365)

# 5. ÎØ∏Îûò ÏäµÎèÑÍ∞íÎèÑ ÎÑ£Ïñ¥Ïïº ÏòàÏ∏° Í∞ÄÎä•
# Ïòà: ÏµúÍ∑º 1ÎÖÑÍ∞Ñ ÏäµÎèÑÍ∞íÏùÑ ÏàúÌôòÌï¥ÏÑú Ï±ÑÏõÄ (Í∞ÑÎã®Ìïú Ïòà)
recent_humidity = df["humidity"].iloc[-365:].values.tolist()
repeat_count = len(future) // len(recent_humidity) + 1
future["humidity"] = (recent_humidity * repeat_count)[:len(future)]

# 6. ÏòàÏ∏° ÏàòÌñâ
forecast = model.predict(future)

# 7. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 5))
plt.plot(forecast["ds"], forecast["yhat"], label="ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò®")
plt.title("Prophet ÏòàÏ∏° Í≤∞Í≥º (ÏäµÎèÑ Ìè¨Ìï®)")
plt.xlabel("ÎÇ†Ïßú")
plt.ylabel("ÌèâÍ∑†Í∏∞Ïò® (¬∞C)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 8. ÏäµÎèÑ ÏòÅÌñ• Î∂ÑÏÑù
model.plot_components(forecast)
```

```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Îç∞Ïù¥ÌÑ∞ Î°úÎî© Î∞è Ï†ÑÏ≤òÎ¶¨
df = pd.read_csv("yeobo.csv")
df["ds"] = pd.to_datetime(df["ÏùºÏãú"])
df["y"] = df["ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"]
df["humidity"] = df["ÌèâÍ∑† ÏÉÅÎåÄÏäµÎèÑ(%)"]
df = df[["ÏßÄÏ†êÎ™Ö", "ds", "y", "humidity"]].dropna()

accuracy_results = []

for station in df["ÏßÄÏ†êÎ™Ö"].unique():
    station_df = df[df["ÏßÄÏ†êÎ™Ö"] == station][["ds", "y", "humidity"]]
    
    # ‚úÖ 2015~2024ÎÖÑ Îç∞Ïù¥ÌÑ∞Îßå ÌïôÏäµ
    train_df = station_df[(station_df["ds"] >= "2015-01-01") & (station_df["ds"] <= "2024-12-31")]
    if len(train_df) < 400:
        continue

    # 1. Î™®Îç∏ ÌïôÏäµ (Ï†ÑÏ≤¥ Í∏∞Í∞Ñ)
    model = Prophet()
    model.add_regressor("humidity")
    model.fit(train_df)

    # 2. ÎØ∏Îûò ÏòàÏ∏°: 2026ÎÖÑ Ï†ÑÏ≤¥
    future = pd.date_range(start="2026-01-01", end="2026-12-31", freq="D")
    future_df = pd.DataFrame({"ds": future})

    # ÏµúÍ∑º ÏäµÎèÑÍ∞í Î∞òÎ≥µ Ï†ÅÏö© (Ïã§Ï†ú ÎØ∏ÎûòÍ∞íÏùÄ Ïïå Ïàò ÏóÜÏúºÎØÄÎ°ú)
    recent_humidity = train_df["humidity"].iloc[-365:].tolist()
    future_df["humidity"] = (recent_humidity * ((len(future_df) // len(recent_humidity)) + 1))[:len(future_df)]

    forecast = model.predict(future_df)

    # 3. üéØ 2026ÎÖÑ 9~11Ïõî ÏòàÏ∏° Í≤∞Í≥ºÎßå Ï∂îÏ∂ú
    fall_forecast = forecast[(forecast["ds"] >= "2026-09-01") & (forecast["ds"] < "2026-12-01")]

    # 4. ÏãúÍ∞ÅÌôî
    plt.figure(figsize=(12, 4))
    plt.plot(fall_forecast["ds"], fall_forecast["yhat"], label="ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò®")
    plt.fill_between(fall_forecast["ds"], fall_forecast["yhat_lower"], fall_forecast["yhat_upper"], alpha=0.3, label="Ïã†Î¢∞ Íµ¨Í∞Ñ")
    plt.title(f"{station} - 2026ÎÖÑ Í∞ÄÏùÑ ÌèâÍ∑†Í∏∞Ïò® ÏòàÏ∏°")
    plt.xlabel("ÎÇ†Ïßú")
    plt.ylabel("ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò® (¬∞C)")
    plt.grid(True)
    plt.legend()
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # 5. üìä Ï†ïÌôïÎèÑ ÌèâÍ∞Ä (2024ÎÖÑ Í∞ÄÏùÑ Îç∞Ïù¥ÌÑ∞Î°ú ÌèâÍ∞Ä)
    actual = station_df[(station_df["ds"] >= "2024-09-01") & (station_df["ds"] < "2024-12-01")]
    eval_future = pd.date_range(start="2024-09-01", end="2024-11-30", freq="D")
    eval_df = pd.DataFrame({"ds": eval_future})
    eval_df["humidity"] = (recent_humidity * 2)[:len(eval_df)]

    eval_forecast = model.predict(eval_df)

    if len(actual) == len(eval_forecast):
        mse = mean_squared_error(actual["y"].values, eval_forecast["yhat"].values)
        r2 = r2_score(actual["y"].values, eval_forecast["yhat"].values)
        accuracy_results.append({
            "ÏßÄÏ†êÎ™Ö": station,
            "MSE (2024 Í∞ÄÏùÑ)": round(mse, 2),
            "R¬≤ (2024 Í∞ÄÏùÑ)": round(r2, 3)
        })

# 6. Ï†ÑÏ≤¥ ÏßÄÏ†ê Ï†ïÌôïÎèÑ Ï∂úÎ†•
print("\nüìä ÏßÄÏ†êÎ≥Ñ Ï†ïÌôïÎèÑ ÌèâÍ∞Ä Í≤∞Í≥º (2024ÎÖÑ Í∞ÄÏùÑ Í∏∞Ï§Ä)")
for acc in accuracy_results:
    print(f"{acc['ÏßÄÏ†êÎ™Ö']}: MSE = {acc['MSE (2024 Í∞ÄÏùÑ)']}, R¬≤ = {acc['R¬≤ (2024 Í∞ÄÏùÑ)']}")
```

```python
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 1. Îç∞Ïù¥ÌÑ∞ Î°úÎî© Î∞è Ï†ïÎ¶¨
df = pd.read_csv("yeobo.csv")
df["ÏùºÏãú"] = pd.to_datetime(df["ÏùºÏãú"])
df = df.sort_values("ÏùºÏãú")
df = df[["ÏßÄÏ†êÎ™Ö", "ÏùºÏãú", "ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"]].dropna()

results = []
accuracy_scores = []

# 2. ÏßÄÏ†êÎ≥Ñ Ï≤òÎ¶¨
for station in df["ÏßÄÏ†êÎ™Ö"].unique():
    station_df = df[df["ÏßÄÏ†êÎ™Ö"] == station][["ÏùºÏãú", "ÌèâÍ∑†Í∏∞Ïò®(¬∞C)"]].copy()
    station_df = station_df.rename(columns={"ÏùºÏãú": "ds", "ÌèâÍ∑†Í∏∞Ïò®(¬∞C)": "y"})

    # ‚úÖ 2-1. ÌïôÏäµ Î≤îÏúÑ: 2015~2024ÎÖÑ
    train_df = station_df[(station_df["ds"] >= "2015-01-01") & (station_df["ds"] <= "2024-12-31")]
    if len(train_df) < 400:
        continue

    model = Prophet()
    model.fit(train_df)

    # ‚úÖ 2-2. ÎØ∏Îûò ÏòàÏ∏° (2026ÎÖÑ)
    future = pd.date_range(start="2026-01-01", end="2026-12-31", freq="D")
    future_df = pd.DataFrame({"ds": future})
    forecast = model.predict(future_df)

    forecast_filtered = forecast[(forecast["ds"] >= "2026-09-01") & (forecast["ds"] < "2026-12-01")]
    forecast_filtered["ÏßÄÏ†êÎ™Ö"] = station
    forecast_filtered = forecast_filtered[["ds", "yhat", "ÏßÄÏ†êÎ™Ö"]].rename(columns={"ds": "ÎÇ†Ïßú", "yhat": "ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò®"})
    results.append(forecast_filtered)

    # ‚úÖ 2-3. Ï†ïÌôïÎèÑ ÌèâÍ∞Ä: 2024ÎÖÑ 9~11Ïõî ÏòàÏ∏° vs Ïã§Ï†ú ÎπÑÍµê
    eval_future = pd.date_range(start="2024-09-01", end="2024-11-30", freq="D")
    eval_df = pd.DataFrame({"ds": eval_future})
    eval_forecast = model.predict(eval_df)

    actual = station_df[(station_df["ds"] >= "2024-09-01") & (station_df["ds"] < "2024-12-01")]

    if len(actual) == len(eval_forecast):
        mse = mean_squared_error(actual["y"].values, eval_forecast["yhat"].values)
        r2 = r2_score(actual["y"].values, eval_forecast["yhat"].values)
        accuracy_scores.append({"ÏßÄÏ†êÎ™Ö": station, "MSE": round(mse, 2), "R2": round(r2, 3)})

# 3. Í≤∞Í≥º Ï∂úÎ†•
forecast_all = pd.concat(results, ignore_index=True)
accuracy_df = pd.DataFrame(accuracy_scores)

print("\nüìÖ [2026ÎÖÑ 9~11Ïõî ÌèâÍ∑†Í∏∞Ïò® ÏòàÏ∏° Í≤∞Í≥º]")
print(forecast_all)

print("\nüìä [ÏßÄÏ†êÎ≥Ñ Prophet ÏòàÏ∏° Ï†ïÌôïÎèÑ ÌèâÍ∞Ä (2024ÎÖÑ Í∞ÄÏùÑ Í∏∞Ï§Ä)]")
print(accuracy_df)

# 4. ÏãúÍ∞ÅÌôî
for station in forecast_all["ÏßÄÏ†êÎ™Ö"].unique():
    data = forecast_all[forecast_all["ÏßÄÏ†êÎ™Ö"] == station]

    plt.figure(figsize=(12, 4))
    plt.rcParams["font.family"] = "Malgun Gothic"
    plt.rcParams['axes.unicode_minus'] = False
    plt.plot(data["ÎÇ†Ïßú"], data["ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò®"], marker='o', linestyle='-', label=f"{station}")
    plt.title(f"2026ÎÖÑ Í∞ÄÏùÑ ÌèâÍ∑†Í∏∞Ïò® ÏòàÏ∏° - {station}")
    plt.xlabel("ÎÇ†Ïßú")
    plt.ylabel("ÏòàÏ∏° ÌèâÍ∑†Í∏∞Ïò® (¬∞C)")
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.tight_layout()
    plt.show()
```
