```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
%config IPCompleter.greedy=True
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
housing = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
print(housing.head(5))
```

```python
from IPython.display import display, HTML

html = housing.to_html()
scroll_html = f"""
<div style='height:400px; overflow:auto;'>  
{html}
</div>
"""

display(HTML(scroll_html))
```

```python
# ZN: 25,000제곱피트 이상의 면적으로 구역 지정된 주거용 토지의 비율
# INDUS: 도시당 비소매업 면적 비율
# CHAS: Charles River 더미 변수(= 해당 구역이 강의 경계를 이루는 경우 1, 그렇지 않은 경우 0)
# NOX: 일산화질소 농도(1000만 분의 1)
# RM: 주택당 평균 방 수
# AGE: 1940년 이전에 건축된 자가 거주 주택의 비율
# DIS: 보스턴 고용 센터 5곳까지의 가중 거리
# RAD: 방사형 고속도로 접근성 지수
# TAX: 10,000달러당 전체 가치 재산세율
# PTRATIO: 도시별 교사-학생 비율 
# B: 1000(Bk−0.63)2 여기서 Bk는 도시별 흑인 비율입니다.  
# LSTAT: 인구의 하위 지위(%)
# MEDV: 수천 달러 단위의 자가 주택의 중간 가치
```

```python
sns.heatmap(housing.corr(), annot= True, fmt=".2f") #피쳐들의 내용을 명확하게 이해하기 어려워 상관계수를 보고 해당 내용을 살펴볼 예졍
```

```python
housing.MEDV.unique() #현재 그래프를 보고 MEDV와 LSTAT가 상관계수가 높아보여 분석읗 해볼려고하는데 데이터 종류가 너무 많음
```

```python
bins = [0, 10, 20, 30, 40, 50] #최솟값이 0이고 최댓값이 50이라서 기준을 이렇게 함
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

housing['MEDV_CLASS'] = pd.cut(housing.MEDV, bins = bins , labels = labels)
```

```python
sns.barplot(x = 'MEDV_CLASS', y = 'LSTAT', data= housing)
plt.rcParams['font.family'] = "Malgun Gothic"
plt.title(" 자가 주택의 중간 가치별 인간의 지위")
# =>주택가치가 낮은 집이 지위가 낮다
```

```python
sns.barplot(x='LSTAT', y='MEDV_CLASS', hue='RM', data=housing)
plt.rcParams['font.family'] = "Malgun Gothic"
plt.title(" 자가 주택의 중간 가치별 인간의 지위")
# 집가치가 높을수록 혹은 지위가 높을수록 방의 갯수가 많아지는 현상이 보임
```

```python
sns.barplot(x= 'MEDV_CLASS', y = 'CHAS', data = housing) #강변가와 집값을 비교했는데 강변가가 집값이 비싸다는 결론이 나옴
```

```python
sns.scatterplot(x='DIS', y='MEDV', data=housing)
plt.title("고용센터와 가격") #고용센터가 가까울수록 가격이 낮음
```

```python
sns.scatterplot(x=housing['CRIM'], y = housing['TAX']) #세금이 600달러 이상의 범죄율이 군집되어 있음
```

```python
housing.groupby('TAX')['PTRATIO'].unique() # 216은 데이터가 두개? 좀 더 자세히 봐야겠음
```

```python
nunique_ptratio_count = housing.groupby('TAX')['PTRATIO'].nunique()
print(nunique_ptratio_count)
tax_multi_ptratio = nunique_ptratio_count[nunique_ptratio_count > 1].index

print(tax_multi_ptratio)
result = housing[housing['TAX'].isin(tax_multi_ptratio)]
html = result.to_html()
scrollable_html = f"""
<div style='height:400px; overflow:auto;'>
{html}
</div>
"""

display(HTML(scrollable_html))
```

```python
key = ['RM', 'LSTAT', 'PTRATIO', 'CRIM', 'TAX']
plt.figure(figsize=(15,10))
for i, feature in enumerate(key, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(x=housing[feature], y=housing['MEDV'])
    plt.title(f'{feature} vs MEDV')
plt.show()
#집값이랑 관계성을 분산도로 확인함
```
