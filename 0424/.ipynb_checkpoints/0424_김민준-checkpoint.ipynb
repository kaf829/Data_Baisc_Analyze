```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
%config IPCompleter.greedy=True
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
housing = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
print(housing.head(5))
```

```python
from IPython.display import display, HTML

html = housing.to_html()
scroll_html = f"""
<div style='height:400px; overflow:auto;'>  
{html}
</div>
"""

display(HTML(scroll_html))
```

```python
bins = [0, 10, 20, 30, 40, 50] # 0  50   
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

housing['MEDV_CLASS'] = pd.cut(housing.MEDV, bins = bins , labels = labels)
```

```python
sns.barplot(x = 'MEDV_CLASS', y = 'LSTAT', data= housing)
plt.rcParams['font.family'] = "Malgun Gothic"
plt.title("      ")
# =>    
```

```python
sns.barplot(x='LSTAT', y='MEDV_CLASS', hue='RM', data=housing)
plt.rcParams['font.family'] = "Malgun Gothic"
plt.title("      ")
#          
```

```python
sns.barplot(x= 'MEDV_CLASS', y = 'CHAS', data = housing) #       
```

```python
nunique_ptratio_count = housing.groupby('TAX')['PTRATIO'].nunique()
print(nunique_ptratio_count)
tax_multi_ptratio = nunique_ptratio_count[nunique_ptratio_count > 1].index

print(tax_multi_ptratio)
result = housing[housing['TAX'].isin(tax_multi_ptratio)]
html = result.to_html()
scrollable_html = f"""
<div style='height:400px; overflow:auto;'>
{html}
</div>
"""

display(HTML(scrollable_html))
```
