```python
import re

def find_M(datas):
    wow = re.search('[A-Z][a-z]+\.', datas)
    if wow == None:
        return None
    else:
        print(wow)
    return wow.group(0)
```

```python
total_set.Name_convert = total_set.Name.map(find_M)
```

```python
import matplotlib.pyplot as plt

labels = ["Survived", "Dead"]
explode = [0.01, 0.1]
colors = ['skyblue', 'salmon']  #   /  

class1_survived = len(df_train[(df_train['Pclass'] == 1) & (df_train['Survived'] == 1)])
class1_dead     = len(df_train[(df_train['Pclass'] == 1) & (df_train['Survived'] == 0)])

class2_survived = len(df_train[(df_train['Pclass'] == 2) & (df_train['Survived'] == 1)])
class2_dead     = len(df_train[(df_train['Pclass'] == 2) & (df_train['Survived'] == 0)])

class3_survived = len(df_train[(df_train['Pclass'] == 3) & (df_train['Survived'] == 1)])
class3_dead     = len(df_train[(df_train['Pclass'] == 3) & (df_train['Survived'] == 0)])

#   
data = [
    (class1_survived, class1_dead),
    (class2_survived, class2_dead),
    (class3_survived, class3_dead)
]

titles = ["1st Class", "2nd Class", "3rd Class"]

fig, axs = plt.subplots(1, 3, figsize=(12, 6))  # 1 3 ,   

for i in range(3):
    axs[i].pie(data[i], labels=labels, autopct='%1.1f%%', startangle=90,
               explode=explode, colors=colors, shadow=True)
    axs[i].set_title(f"{titles[i]} Survival Rate")

plt.tight_layout()
plt.show()
```
