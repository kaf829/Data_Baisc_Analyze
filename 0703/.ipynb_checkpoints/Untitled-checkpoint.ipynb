```python
df = pd.read_csv("daily-total-female-births-CA.csv")
df.head()
```

```python
timestap = df.births
```

```python
def SMA(array, window):
    array = np.array(array)
    assert array.ndim == 1, "1차원 array만 입력할 수 있습니다"
    n = len(array)

    result = np.empty(n)
    result[:] = np.nan

    ma = []
    for i in range(0 , n-window+1):
        m = np.mean(array[i: i+window])
        ma.append(m)

    result[window-1 : ] = ma

    return np.array(result)
```

```python
window = 10

sma = SMA(timestap, window)

timestap.plot(figsize = (15,6))
plt.plot(sma)
plt.legend(['value', 'ma'])
plt.title("simple Moving Average with window sie = {}".format(window))
plt.show()
```

```python
import pandas as pd
import numpy as np

data = {'A': [1,2, np.nan, 4,5],
        'B': [5,np.nan, np.nan, 8,9],
        'C': [9,10,11,12,13]}
df = pd.DataFrame(data)
print(df)
```

```python
df_interpolated = df.interpolate()
print(df_interpolated)
```

```python
df_interpolated2 = df['A'].interpolate()
print(df_interpolated2)
```

```python
df_interpolated= df.interpolate(method='polynomial' , order =1)
print(df_interpolated)
```

```python
df_interpolated = df.interpolate(method = 'linear', limit = 1, limit_direction='backward')
print(df_interpolated)
```

```python
df_interpolated = df.interpolate(method = 'slinear', limit = 1, limit_direction='backward')
print(df_interpolated)
```

```python
df = pd.read_csv('AAPL.csv')
df.head()
```

```python
df['ds'] = df['Date']
df['y'] = df['Close']
data = df[['ds', 'y']]
data = data[-1000:]
data.head()
```

```python
from prophet import Prophet

prophet_m = Prophet().fit(data)
```

```python
future = prophet_m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = prophet_m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
prophet_m.plot(forecast)
```

```python
path = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-car-sales.csv"
df = pd.read_csv(path, header = 0)
print(df.shape)
print(df.head())
```

```python
from matplotlib import pyplot

df.plot()
pyplot.show()
```

```python
df.columns = ['ds', 'y']
df['ds'] = pd.to_datetime(df['ds'])
```

```python
df.tail()
```

```python
model = Prophet()
model.fit(df)
```

```python
import pandas as pd

last_year = list()
for i in range(1, 13):
    last_year.append(['1968-%02d' % i])

last_year = pd.DataFrame(last_year, columns=['ds'])
last_year['ds'] = pd.to_datetime(last_year['ds'])
print(last_year)
```

```python
forecast = model.predict(last_year)
```

```python
model.plot(forecast)
```
