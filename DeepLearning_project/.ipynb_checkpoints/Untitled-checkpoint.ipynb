```python
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt

# 결측 보간 먼저 (원본 보존 필요 시 따로 백업)
df_filled = df_raw.copy()
df_filled['Global_active_power'] = df_filled['Global_active_power'].interpolate(method='time')

# 학습용 시계열만 추출 (1차원)
X = df_filled[['Global_active_power']].dropna()

# 모델 정의 및 학습
iso_model = IsolationForest(contamination=0.01, random_state=42)
X['anomaly'] = iso_model.fit_predict(X)

# 이상치는 -1, 정상은 1 → 1: 정상 / -1: 이상 → 0/1 로 변경
X['anomaly'] = X['anomaly'].map({1: 0, -1: 1})

# 시각화
plt.figure(figsize=(20,5))
plt.plot(X.index, X['Global_active_power'], label='Power Usage', color='black')
plt.scatter(X.index[X['anomaly'] == 1], X['Global_active_power'][X['anomaly'] == 1],
            color='red', label='Anomaly', s=10)
plt.title('Anomaly Detection using Isolation Forest')
plt.ylabel('kW')
plt.xlabel('Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. 시간당 평균 전력 소비량
df_hourly = df_filled['Global_active_power'].resample('H').mean().dropna()

# 2. 이동 평균 및 이상치 기준 계산 (24시간 기준)
rolling_mean = df_hourly.rolling(window=24).mean()
residual = df_hourly - rolling_mean
threshold = 3 * residual.std()

# 3. 이상치 마스크 (이동평균에서 3σ 이상 벗어난 값)
anomaly_mask = residual.abs() > threshold
plt.rcParams['font.family'] = 'Malgun Gothic'
# 4. 시각화
plt.figure(figsize=(20, 5))
plt.plot(df_hourly, label='전력량', color='gray')
plt.plot(rolling_mean, label='이동평균', color='blue', linestyle='--')
plt.scatter(df_hourly.index[anomaly_mask], df_hourly[anomaly_mask],
            color='red', label='이상치', s=15)
plt.xlabel('Time')
plt.ylabel('kW')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.optimizers import Adam, RMSprop
from keras.callbacks import EarlyStopping
from keras.backend import clear_session
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# 시퀀스 생성 함수
def create_dataset(data, look_back):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i+look_back])
        y.append(data[i+look_back])
    return np.array(X), np.array(y)

# 하이퍼파라미터 정의
look_backs = [1440] # 1,60 일시 삭제
units_list = [32, 64, 128]
batch_sizes = [32, 64]
epochs_list = [10, 30, 50]
dropouts = [0.0, 0.25, 0.5]
optimizers = ['adam', 'rmsprop']
layer_options = [1, 2, 3]

# 시계열 데이터 정규화 및 분할
df_minutely = df_cleaned.resample('min').mean().dropna()
scaler = MinMaxScaler()
series = df_minutely.values.reshape(-1, 1)
scaled_series = scaler.fit_transform(series)

total_len = len(scaled_series)
train_size = int(total_len * 0.6)
val_size = int(total_len * 0.2)

train_data = scaled_series[:train_size]
val_data = scaled_series[train_size:train_size+val_size]
test_data = scaled_series[train_size+val_size:]

results = []

# 실험 루프
for look_back in look_backs:
    X_train, y_train = create_dataset(train_data, look_back)
    X_val, y_val = create_dataset(val_data, look_back)
    X_test, y_test = create_dataset(test_data, look_back)

    X_train = X_train.reshape(-1, look_back, 1)
    X_val = X_val.reshape(-1, look_back, 1)
    X_test = X_test.reshape(-1, look_back, 1)
    i = 1
    for units in units_list:
        for batch_size in batch_sizes:
            for epochs in epochs_list:
                for dropout in dropouts:
                    for opt in optimizers:
                        for layers in layer_options:
                            clear_session()

                            print(f"⏳ Training LSTM: look_back={look_back}, units={units}, batch={batch_size}, "
                                  f"epochs={epochs}, dropout={dropout}, opt={opt}, layers={layers}")

                            model = Sequential()
                            if layers == 1:
                                model.add(LSTM(units, input_shape=(look_back, 1)))
                            elif layers == 2:
                                model.add(LSTM(units, return_sequences=True, input_shape=(look_back, 1)))
                                model.add(LSTM(units // 2))
                            elif layers == 3:
                                model.add(LSTM(units, return_sequences=True, input_shape=(look_back, 1)))
                                model.add(LSTM(units, return_sequences=True))
                                model.add(LSTM(units // 2))

                            if dropout > 0.0:
                                model.add(Dropout(dropout))
                            model.add(Dense(1))

                            model.compile(optimizer=opt, loss='mse')
                            early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

                            model.fit(X_train, y_train,
                                      epochs=epochs,
                                      batch_size=batch_size,
                                      validation_data=(X_val, y_val),
                                      callbacks=[early_stop],
                                      verbose=0)

                            pred = model.predict(X_test)
                            pred_inv = scaler.inverse_transform(pred)
                            y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
                            mae = mean_absolute_error(y_test_inv, pred_inv)
                            i +=1
                            print(f"✅{i}번째 Done → MAE: {mae:.4f}")
                            
                            results.append({
                                'look_back': look_back,
                                'units': units,
                                'batch_size': batch_size,
                                'epochs': epochs,
                                'dropout': dropout,
                                'optimizer': opt,
                                'layers': layers,
                                'MAE': mae
                            })

# 결과 정렬
results_df = pd.DataFrame(results).sort_values(by='MAE')``
print(results_df.head(5))
```

```python
def create_dataset(data, look_back):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i+look_back])
        y.append(data[i+look_back])
    return np.array(X), np.array(y)
```

```python
# 이전 상태 초기화 후 다시 실행
import pandas as pd
from itertools import product

# 마지막 완료된 조합
last_completed = {
    'look_back': 1440,
    'units': 32,
    'batch_size': 32,
    'epochs': 10,
    'dropout': 0.0,
    'optimizer': 'adam',
    'layers': 2
}

# 가능한 모든 조합
look_backs = [1, 60, 1440]
units_list = [32, 64, 128]
batch_sizes = [32, 64]
epochs_list = [10, 30, 50]
dropouts = [0.0, 0.25, 0.5]
optimizers = ['adam', 'rmsprop']
layer_options = [1, 2, 3]

# 전체 조합 리스트 생성
param_grid = list(product(
    look_backs,
    units_list,
    batch_sizes,
    epochs_list,
    dropouts,
    optimizers,
    layer_options
))

# DataFrame으로 변환
param_df = pd.DataFrame(param_grid, columns=[
    'look_back', 'units', 'batch_size', 'epochs', 'dropout', 'optimizer', 'layers'
])

# 마지막 완료된 조합 위치 찾기
idx_last = param_df[
    (param_df['look_back'] == last_completed['look_back']) &
    (param_df['units'] == last_completed['units']) &
    (param_df['batch_size'] == last_completed['batch_size']) &
    (param_df['epochs'] == last_completed['epochs']) &
    (param_df['dropout'] == last_completed['dropout']) &
    (param_df['optimizer'] == last_completed['optimizer']) &
    (param_df['layers'] == last_completed['layers'])
].index[0]

# 이후 조합만 추출
remaining_param_df = param_df.iloc[idx_last + 1:].reset_index(drop=True)

```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.optimizers import Adam, RMSprop
from keras.callbacks import EarlyStopping
from keras.backend import clear_session
from sklearn.metrics import mean_absolute_error
import numpy as np

# 실험 루프 (남은 조합만 반복)
for i, row in remaining_param_df.iterrows():
    look_back = row['look_back']
    units = row['units']
    batch_size = row['batch_size']
    epochs = row['epochs']
    dropout = row['dropout']
    optimizer = row['optimizer']
    layers = row['layers']

    X_train, y_train = create_dataset(train_data, look_back)
    X_val, y_val = create_dataset(val_data, look_back)
    X_test, y_test = create_dataset(test_data, look_back)

    X_train = X_train.reshape(-1, look_back, 1)
    X_val = X_val.reshape(-1, look_back, 1)
    X_test = X_test.reshape(-1, look_back, 1)

    clear_session()

    print(f"⏳ Training LSTM: look_back={look_back}, units={units}, batch={batch_size}, "
          f"epochs={epochs}, dropout={dropout}, opt={optimizer}, layers={layers}")

    # 모델 구성
    model = Sequential()
    if layers == 1:
        model.add(LSTM(units, input_shape=(look_back, 1)))
    elif layers == 2:
        model.add(LSTM(units, return_sequences=True, input_shape=(look_back, 1)))
        model.add(LSTM(units // 2))
    elif layers == 3:
        model.add(LSTM(units, return_sequences=True, input_shape=(look_back, 1)))
        model.add(LSTM(units, return_sequences=True))
        model.add(LSTM(units // 2))

    if dropout > 0.0:
        model.add(Dropout(dropout))
    model.add(Dense(1))

    model.compile(optimizer=optimizer, loss='mse')

    early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

    model.fit(X_train, y_train,
              epochs=epochs,
              batch_size=batch_size,
              validation_data=(X_val, y_val),
              callbacks=[early_stop],
              verbose=0)

    pred = model.predict(X_test)
    pred_inv = scaler.inverse_transform(pred)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
    mae = mean_absolute_error(y_test_inv, pred_inv)

    print(f"✅ Done → MAE: {mae:.4f}")

    results.append({
        'look_back': look_back,
        'units': units,
        'batch_size': batch_size,
        'epochs': epochs,
        'dropout': dropout,
        'optimizer': optimizer,
        'layers': layers,
        'MAE': mae
    })
```

```python
from keras.models import Sequential
from keras.layers import LSTM, GRU, Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from keras.optimizers import Adam, RMSprop
from keras.callbacks import EarlyStopping
from keras.backend import clear_session
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# 시퀀스 생성 함수
def create_dataset(data, look_back):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i+look_back])
        y.append(data[i+look_back])
    return np.array(X), np.array(y)

# 하이퍼파라미터 정의
look_backs = [1, 60]
units_list = [32, 64, 128]
batch_sizes = [32, 64]
epochs_list = [10, 30, 50]
dropouts = [0.0, 0.25, 0.5]
optimizers = ['adam', 'rmsprop']
layer_options = [1, 2, 3]
model_types = ['GRU', 'CNN', 'DENSE']

# 시계열 데이터 정규화 및 분할
df_minutely = df_cleaned.resample('min').mean().dropna()
scaler = MinMaxScaler()
series = df_minutely.values.reshape(-1, 1)
scaled_series = scaler.fit_transform(series)

total_len = len(scaled_series)
train_size = int(total_len * 0.6)
val_size = int(total_len * 0.2)

train_data = scaled_series[:train_size]
val_data = scaled_series[train_size:train_size+val_size]
test_data = scaled_series[train_size+val_size:]

results = []
i = 1
# 실험 루프
for model_type in model_types:
    for look_back in look_backs:
        X_train, y_train = create_dataset(train_data, look_back)
        X_val, y_val = create_dataset(val_data, look_back)
        X_test, y_test = create_dataset(test_data, look_back)

        X_train = X_train.reshape(-1, look_back, 1)
        X_val = X_val.reshape(-1, look_back, 1)
        X_test = X_test.reshape(-1, look_back, 1)
        
        for units in units_list:
            for batch_size in batch_sizes:
                for epochs in epochs_list:
                    for dropout in dropouts:
                        for opt in optimizers:
                            for layers in layer_options:
                                clear_session()
                                
                                print(f"⏳{i}번째 Training {model_type}: look_back={look_back}, units={units}, batch={batch_size}, "
                                      f"epochs={epochs}, dropout={dropout}, opt={opt}, layers={layers}")

                                model = Sequential()

                                # ✅ 모델별 구조 분기
                                if model_type == 'LSTM':
                                    if layers == 1:
                                        model.add(LSTM(units, input_shape=(look_back, 1)))
                                    elif layers == 2:
                                        model.add(LSTM(units, return_sequences=True, input_shape=(look_back, 1)))
                                        model.add(LSTM(units // 2))
                                    elif layers == 3:
                                        model.add(LSTM(units, return_sequences=True, input_shape=(look_back, 1)))
                                        model.add(LSTM(units, return_sequences=True))
                                        model.add(LSTM(units // 2))
                                elif model_type == 'GRU':
                                    if layers == 1:
                                        model.add(GRU(units, input_shape=(look_back, 1)))
                                    elif layers == 2:
                                        model.add(GRU(units, return_sequences=True, input_shape=(look_back, 1)))
                                        model.add(GRU(units // 2))
                                    elif layers == 3:
                                        model.add(GRU(units, return_sequences=True, input_shape=(look_back, 1)))
                                        model.add(GRU(units, return_sequences=True))
                                        model.add(GRU(units // 2))
                                elif model_type == 'CNN':
                                    model.add(Conv1D(filters=units, kernel_size=3, activation='relu', input_shape=(look_back, 1)))
                                    model.add(MaxPooling1D(pool_size=2))
                                    if layers >= 2:
                                        model.add(Conv1D(filters=units//2, kernel_size=3, activation='relu'))
                                        model.add(MaxPooling1D(pool_size=2))
                                    if layers == 3:
                                        model.add(Conv1D(filters=units//4, kernel_size=3, activation='relu'))
                                    model.add(Flatten())
                                elif model_type == 'DENSE':
                                    model.add(Flatten(input_shape=(look_back, 1)))
                                    for _ in range(layers):
                                        model.add(Dense(units, activation='relu'))

                                # 공통 구성
                                if dropout > 0.0:
                                    model.add(Dropout(dropout))
                                model.add(Dense(1))

                                model.compile(optimizer=opt, loss='mse')
                                early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

                                model.fit(X_train, y_train,
                                          epochs=epochs,
                                          batch_size=batch_size,
                                          validation_data=(X_val, y_val),
                                          callbacks=[early_stop],
                                          verbose=0)

                                pred = model.predict(X_test)
                                pred_inv = scaler.inverse_transform(pred)
                                y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
                                mae = mean_absolute_error(y_test_inv, pred_inv)
                                print(f"✅ Done → MAE: {mae:.4f}")
                                i +=1
                                results.append({
                                    'model': model_type,
                                    'look_back': look_back,
                                    'units': units,
                                    'batch_size': batch_size,
                                    'epochs': epochs,
                                    'dropout': dropout,
                                    'optimizer': opt,
                                    'layers': layers,
                                    'MAE': mae
                                })

# 결과 정렬
results_df = pd.DataFrame(results).sort_values(by='MAE')
print(results_df.head(10))
```

```python
# 전체 통합 코드: 603번째 실험부터 남은 조합 반복 실행
from keras.models import Sequential
from keras.layers import GRU, Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from keras.optimizers import Adam, RMSprop
from keras.callbacks import EarlyStopping
from keras.backend import clear_session
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# 시퀀스 생성 함수
def create_dataset(data, look_back):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back])
        y.append(data[i + look_back])
    return np.array(X), np.array(y)

# 하이퍼파라미터 정의
look_backs = [1, 60]
units_list = [32, 64, 128]
batch_sizes = [32, 64]
epochs_list = [10, 30, 50]
dropouts = [0.0, 0.25, 0.5]
optimizers = ['adam', 'rmsprop']
layer_options = [1, 2, 3]
model_types = ['GRU', 'CNN', 'DENSE']

# 데이터 준비
df_minutely = df_cleaned.resample('min').mean().dropna()
scaler = MinMaxScaler()
series = df_minutely.values.reshape(-1, 1)
scaled_series = scaler.fit_transform(series)

total_len = len(scaled_series)
train_size = int(total_len * 0.6)
val_size = int(total_len * 0.2)

train_data = scaled_series[:train_size]
val_data = scaled_series[train_size:train_size + val_size]
test_data = scaled_series[train_size + val_size:]

# 조합 생성 및 필터링
from itertools import product
all_combinations = list(product(model_types, look_backs, units_list, batch_sizes, epochs_list, dropouts, optimizers, layer_options))
remaining_combinations = all_combinations[602:]  # 603번째부터 실행

results = []
i = 603  # 시작 인덱스

for model_type, look_back, units, batch_size, epochs, dropout, opt, layers in remaining_combinations:
    clear_session()

    print(f"⏳{i}번째 Training {model_type}: look_back={look_back}, units={units}, batch={batch_size}, "
          f"epochs={epochs}, dropout={dropout}, opt={opt}, layers={layers}")

    X_train, y_train = create_dataset(train_data, look_back)
    X_val, y_val = create_dataset(val_data, look_back)
    X_test, y_test = create_dataset(test_data, look_back)

    X_train = X_train.reshape(-1, look_back, 1)
    X_val = X_val.reshape(-1, look_back, 1)
    X_test = X_test.reshape(-1, look_back, 1)

    model = Sequential()

    if model_type == 'GRU':
        if layers == 1:
            model.add(GRU(units, input_shape=(look_back, 1)))
        elif layers == 2:
            model.add(GRU(units, return_sequences=True, input_shape=(look_back, 1)))
            model.add(GRU(units // 2))
        elif layers == 3:
            model.add(GRU(units, return_sequences=True, input_shape=(look_back, 1)))
            model.add(GRU(units, return_sequences=True))
            model.add(GRU(units // 2))
    elif model_type == 'CNN':
        model.add(Conv1D(filters=units, kernel_size=3, activation='relu', input_shape=(look_back, 1)))
        model.add(MaxPooling1D(pool_size=2))
        if layers >= 2:
            model.add(Conv1D(filters=units // 2, kernel_size=3, activation='relu'))
            model.add(MaxPooling1D(pool_size=2))
        if layers == 3:
            model.add(Conv1D(filters=units // 4, kernel_size=3, activation='relu'))
        model.add(Flatten())
    elif model_type == 'DENSE':
        model.add(Flatten(input_shape=(look_back, 1)))
        for _ in range(layers):
            model.add(Dense(units, activation='relu'))

    if dropout > 0.0:
        model.add(Dropout(dropout))
    model.add(Dense(1))

    model.compile(optimizer=opt, loss='mse')
    early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

    model.fit(X_train, y_train,
              epochs=epochs,
              batch_size=batch_size,
              validation_data=(X_val, y_val),
              callbacks=[early_stop],
              verbose=0)

    pred = model.predict(X_test)
    pred_inv = scaler.inverse_transform(pred)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
    mae = mean_absolute_error(y_test_inv, pred_inv)

    print(f"✅ Done → MAE: {mae:.4f}")
    results.append({
        'model': model_type,
        'look_back': look_back,
        'units': units,
        'batch_size': batch_size,
        'epochs': epochs,
        'dropout': dropout,
        'optimizer': opt,
        'layers': layers,
        'MAE': mae
    })

    i += 1

# 결과 정렬
results_df = pd.DataFrame(results).sort_values(by='MAE')
results_df.head(10)
```

```python
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd
import numpy as np
from itertools import product

# Prophet 입력 형식으로 변환
df_prophet = df_cleaned.reset_index()
df_prophet.columns = ['ds', 'y']

# 데이터 분할 (train 60%, val 20%, test 20%)
total_len = len(df_prophet)
train_end = int(total_len * 0.6)
val_end = int(total_len * 0.8)

train = df_prophet[:train_end]
val = df_prophet[train_end:val_end]
test = df_prophet[val_end:]

# 하이퍼파라미터 후보 정의
changepoint_prior_scales = [0.01, 0.1]
seasonality_prior_scales = [1.0, 10.0]
seasonality_modes = ['additive', 'multiplicative']
yearly_opts = [True, False]
weekly_opts = [True, False]
daily_opts = [True, False]

# 결과 저장 리스트
results = []

# 그리드 서치 실행
for cps, sps, sm, y_opt, w_opt, d_opt in product(
    changepoint_prior_scales,
    seasonality_prior_scales,
    seasonality_modes,
    yearly_opts,
    weekly_opts,
    daily_opts
):
    try:
        model = Prophet(
            changepoint_prior_scale=cps,
            seasonality_prior_scale=sps,
            seasonality_mode=sm,
            yearly_seasonality=y_opt,
            weekly_seasonality=w_opt,
            daily_seasonality=d_opt
        )

        model.fit(train)
        future = model.make_future_dataframe(periods=len(val), freq='min')  # 분 단위 예측
        forecast = model.predict(future)

        # 병합을 통한 시간 정렬 (분 단위 불일치 해결)
        merged = pd.merge(val, forecast[['ds', 'yhat']], on='ds', how='inner')

        # 성능 평가
        mae = mean_absolute_error(merged['y'], merged['yhat'])
        rmse = np.sqrt(mean_squared_error(merged['y'], merged['yhat']))

        results.append({
            'changepoint_prior_scale': cps,
            'seasonality_prior_scale': sps,
            'seasonality_mode': sm,
            'yearly': y_opt,
            'weekly': w_opt,
            'daily': d_opt,
            'MAE': mae,
            'RMSE': rmse
        })

    except Exception as e:
        print(f"에러 발생: {e}")
        continue

# 결과 정리 및 정렬
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='MAE').reset_index(drop=True)

```

```python
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.callbacks import EarlyStopping

# 지정된 파라미터
look_back = 60
units = 128
batch_size = 32
epochs = 50
dropout = 0.25
optimizer = 'adam'
layers = 2

# 시퀀스 데이터 준비
X_train, y_train = create_dataset(train_data, look_back)
X_val, y_val = create_dataset(val_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)

X_train = X_train.reshape(-1, look_back, 1)
X_val = X_val.reshape(-1, look_back, 1)
X_test = X_test.reshape(-1, look_back, 1)

# 모델 구성
model = Sequential()
model.add(LSTM(units, return_sequences=True, input_shape=(look_back, 1)))
model.add(LSTM(units // 2))
model.add(Dropout(dropout))
model.add(Dense(1))

model.compile(optimizer=optimizer, loss='mse')
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# 학습
model.fit(
    X_train, y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_data=(X_val, y_val),
    callbacks=[early_stop],
    verbose=0
)

y_pred = model.predict(X_test)
y_pred_inv = scaler.inverse_transform(y_pred)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

plt.figure(figsize=(15, 5))
plt.plot(y_test_inv, label='True Values', color='black')
plt.plot(y_pred_inv, label='Predicted Values', color='orange')
plt.title('LSTM 예측 결과 (실제값 vs 예측값)')
plt.xlabel('Time Step')
plt.ylabel('Power (kW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
