```python
import matplotlib.pyplot as plt

def boxplot_iris(feature_names, dataset):
    i = 1
    plt.figure(figsize=(11, 9))  # 전체 도화지 크기 설정 (가로 11, 세로 9)

    for col in feature_names:
        plt.subplot(2, 2, i)  # 2행 2열 중 i번째 subplot에 그리기
        plt.axis("on")  # 축 보이도록 설정

        # 축 눈금 설정 (눈금은 보이되, 라벨은 안 보이게)
        plt.tick_params(
            axis="both",
            left=True, right=False, top=False, bottom=True,  # 눈금선
            labelleft=False, labelright=False, labeltop=False, labelbottom=False  # 라벨
        )

        dataset[col].plot(kind="box")  # ✅ data.set → dataset (오타 수정)
        plt.title(col)  # 박스플롯 제목 설정

        i += 1  # 다음 subplot으로 이동

    plt.tight_layout()  # subplot 간격 자동 조절
    plt.show()  # 전체 그림 출력

boxplot_iris(iris.columns[:-1], iris)
```

```python
import seaborn as sns

sns.pairplot(iris, hue = "species_name")
plt.show()
```

```python
import matplotlib.pyplot as plt

def piechart_iris(colName, dataset):
    plt.figure(figsize=(4, 3))  # 도넛 차트 전체 크기 설정

    labels = []  # 범주(예: 0,1,2 또는 setosa 등)
    sizes = []   # 각 범주의 개수

    df = dataset.groupby(colName).size()  # ✅ datasets → dataset (오타 수정)

    for key in df.keys():
        labels.append(key)
        sizes.append(df[key])  # 각 범주 개수 수집


    plt.pie(sizes, labels=labels, autopct="%1.1f%%", shadow=True, startangle=140)
    plt.axis("equal")  # 원형 비율 유지
    plt.title(f"Pie Chart of {colName}")
    plt.show()


piechart_iris("species_name", iris)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,:-2], iris.iloc[:, -1], test_size = 0.33, random_state =42)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion = "gini", splitter ="best", max_depth = None, min_samples_split = 2, min_samples_leaf = 1, random_state =42, max_leaf_nodes = None)
print(model)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

# KFold 설정: 10개 조각으로 나누고 셔플함
cv = KFold(n_splits=10, shuffle=True, random_state=42)

# 교차검증 수행 (X_train, y_train에 대해 10번 학습/검증 반복)
results = cross_val_score(model, X_train, y_train, cv=cv)

# 평균 정확도 계산
fin_result = np.mean(results)

# 개별 fold 결과 출력
for i, score in enumerate(results):
    print(f"{i+1}번째 교차검증 정확도: {score:.4f}")

# 전체 평균 정확도 출력
print(f"\n교차검증 최종 평균 정확도: {fin_result:.4f}")
```

```python
# 결정 트리 분류기와 교차검증 도구 불러오기
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, KFold

# ① 사용할 기본 결정 트리 모델 설정 (아직 하이퍼파라미터는 기본값)
estimator = DecisionTreeClassifier()

# ② KFold 교차검증 설정: 데이터를 10개로 나누고, 무작위 셔플 후 반복 학습
cv = KFold(n_splits=10, random_state=42, shuffle=True)

# ③ 실험할 하이퍼파라미터 목록 정의 (모든 조합을 실험)
parameters = {
    "max_depth": [4, 5, 6, 10],                 # 트리 최대 깊이
    "criterion": ["gini", "entropy"],          # 분할 기준 (지니 불순도 또는 정보 이득)
    "splitter": ["best", "random"],            # 분할 방법 (최선 또는 무작위)
    "max_leaf_nodes": [1, 2, 3, 4, 5],          # 리프 노드 최대 개수
    "random_state": [7, 23, 42, 78],            # 시드 (트리 내부 분할에도 랜덤성이 있음)
    "min_samples_split": [2, 3, 4],             # 내부 노드를 나누기 위한 최소 샘플 수
    "min_samples_leaf": [1, 2, 3, 4, 5]         # 리프노드가 되기 위한 최소 샘플 수
}

# ④ 그리드서치 설정: 모든 하이퍼파라미터 조합을 탐색하고, 교차검증으로 성능 측정
model = GridSearchCV(
    estimator=estimator,           # 기본 모델
    param_grid=parameters,         # 실험할 하이퍼파라미터 조합
    cv=cv,                         # 10폴드 교차검증 설정
    verbose=1,                     # 진행 과정을 출력 (1은 적당한 수준)
    n_jobs=1,                      # 병렬 처리 안 함 (1개 CPU 사용)
    refit=True                     # 최적 파라미터로 다시 전체 학습
)

# ⑤ 모델 훈련 실행: 모든 조합을 시도하여 최고의 조합과 모델을 찾음
model.fit(X_train, y_train)
```

```python
print("best Estimator \n", model.best_estimator_)
print()
print("Best Params \n",model.best_params_);
print()
print("Best score; \n",model.best_score_);
print()
```

```python
from sklearn.metrics import accuracy_score

pred = model.predict(X_test)
acc = accuracy_score(y_test, pred)
print(acc)
```

```python
from sklearn.metrics import confusion_matrix
confMatrix = confusion_matrix(y_test, pred)
print("Confusion Matrix : \n" , confMatrix)
```
