```python
[문제] 타이타닉 데이터 이용(./titanic/train.csv 파일 사용)하여 아래 내용을 만족하는 각각의 데이터 프레임을 만들어 출력하는 예제를 완성하시오.

(1) 결측값 처리
    - Age : 평균값으로 결측값 처리
    - Embarked : 최빈값으로 결측값 처리
(2) 레이블 인코딩
    - Sex
    - Embarked
(3) 원핫인코딩
    - Sex
    - Embarked
```

```python
import pandas as pd
```

```python
titanic = pd.read_csv("./titanic/train.csv")
titanic.head()
```

```python
#1번
from sklearn.impute import SimpleImputer
import numpy as np

imp_mean = SimpleImputer(missing_values = np.nan, strategy = "mean" )
imp_mean.fit_transform(titanic[["Age"]])

age = pd.DataFrame(
    imp_mean.fit_transform(titanic[["Age"]]),
    columns=["Age"]
)
print(age)

imp_most_freq = SimpleImputer(missing_values = np.nan, strategy = "most_frequent" )
imp_most_freq.fit_transform(titanic[["Embarked"]])

embarked = pd.DataFrame(
    imp_most_freq.fit_transform(titanic[["Embarked"]]),
    columns=["Embarked"]
)

print(embarked)
```

```python
#2번 레이블인코딩("SEX")
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(titanic[["Sex"]])
sex = le.transform(titanic[["Sex"]])
label_sex = pd.DataFrame(le.transform(titanic[["Sex"]]), columns=["sex"])
print(label_sex)
```

```python
#2번 레이블 인코딩(embarked)
le = LabelEncoder()
le.fit(titanic[["Embarked"]])
Embarked = le.transform(titanic[["Embarked"]])
label_embarked = pd.DataFrame(Embarked, columns=["Embarked"])
print(label_embarked)
```

```python
#3번 원핫인코딩(embakred)
embarked = pd.get_dummies(titanic[["Embarked"]])
embarked
```

```python
#3번 원핫인코딩(sex)
sex = pd.get_dummies(titanic[["Sex"]])
sex
```
